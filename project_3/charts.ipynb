```python
%matplotlib inline

import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.2)
df_qlearning = pd.read_csv('log_qlearning2.csv')
df_friendq = pd.read_csv('log_friendq1.csv')
df_foeq = pd.read_csv('log_foeq3.csv')
df_uceq = pd.read_csv('log_uceq4.csv')

def format_axis(axis, title):
    axis.set_ylim(0, 0.5)
    axis.set_xlim(0, 10e5)
    axis.set_xlabel('Simulation iteration')
    axis.set_ylabel('Q-value difference')
    axis.set_title(title)
    axis.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    axis.xaxis.major.formatter._useMathText = True
    _ = axis.xaxis.set_ticks(np.arange(0, 10e5 + 1, 1e5))
    _ = axis.yaxis.set_ticks(np.arange(0, 0.51, 0.05))
    axis.grid(b=True, which='major', color='#cccccc', linewidth=0.4, linestyle='--')
    axis.set_facecolor('#f9f9f9')
    plt.setp(axis.spines.values(), color='#cccccc')
    plt.setp([axis.get_xticklines(), axis.get_yticklines()], color='#cccccc')
    
ax4.plot(df_qlearning['timestep'], df_qlearning['error'], linewidth=0.2)
format_axis(ax4, 'Q-learner')

ax3.plot(df_friendq['timestep'], df_friendq['error'], linewidth=0.6)
format_axis(ax3, 'Friend-Q')

ax2.plot(df_foeq['timestep'], df_foeq['error'], linewidth=0.4)
format_axis(ax2, 'Foe-Q')

ax1.plot(df_uceq['timestep'], df_uceq['error'], linewidth=0.6)
format_axis(ax1, 'uCE-Q')

plt.savefig('paper/images/figure2.png')
```
